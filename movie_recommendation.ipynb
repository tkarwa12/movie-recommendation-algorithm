## Data Exploration and Clean Up - 80%

```python
df_credits= pd.read_csv('credits.csv')
df_movies= pd.read_csv('movies.csv')
```

```python
df_credits.head(4)
```

```python
df_movies.head(4)
```

```python
df_movies_credits=df_movies.merge(df_credits, on='title')
```

```python
df_movies_credits.head(4)
```

```python
df_movies_select=df_movies_credits[['movie_id','title','overview','genres','keywords','cast','crew']]
df_movies_select
```

```python
df_movies_select.isnull().sum()
```

```python
df_movies_select.dropna(inplace=True)
```

```python
df_movies_select.duplicated().sum() #number of duplicate rows
```

```python
def convert(obj):#extracts the values from columns genre, cast, and keyword
    L=[]
    for i in ast.literal_eval(obj): #navigates the complex data structure with ease
        L.append(i['name'])
    return L
```

```python
df_movies_select['genres']=df_movies_select['genres'].apply(convert)
df_movies_select['keywords']=df_movies_select['keywords'].apply(convert)
df_movies_select['cast']=df_movies_select['cast'].apply(convert)
df_movies_select['crew']=df_movies_select['crew'].apply(convert)
df_movies_select
```

```python
def get_director(obj):#extracts the director title from crew
    L=[]
    for i in ast.literal_eval(obj): #navigates the complex data structure with ease
        if i['job']=='Director':
            L.append(i['name'])
    return L
```

```python
df_movies_select['overview']=df_movies_select['overview'].apply(lambda x: x.split())
df_movies_select
```

```python
df_movies_select['genres'] = df_movies_select['genres'].apply(lambda x: [i.replace(' ', '') for i in x])
df_movies_select['cast'] = df_movies_select['cast'].apply(lambda x: [i.replace(' ', '') for i in x])
df_movies_select['crew'] = df_movies_select['crew'].apply(lambda x: [i.replace(' ', '') for i in x])
df_movies_select['keywords'] = df_movies_select['keywords'].apply(lambda x: [i.replace(' ', '') for i in x])
df_movies_select
```

```python
df_movies_select['tags']=df_movies_select['overview']+df_movies_select['keywords']+df_movies_select['genres']+df_movies_select['cast']+df_movies_select['crew']
df_movies_select
```

```python
df_movies_tags=df_movies_select[['movie_id','title','tags']]
df_movies_tags
```

```python
#remove brackets from tags columns
df_movies_tags['tags']=df_movies_tags['tags'].apply(lambda x: ' '.join(x))
df_movies_tags
```

```python
df_movies_tags['tags']=df_movies_tags['tags'].apply(lambda x: x.lower())
df_movies_tags
```

## Data Science (Recommendation System) - 20%

```python
from sklearn.feature_extraction.text import CountVectorizer #converts text to a vector

import nltk
from nltk.stem.porter import PorterStemmer #finds the stem of a word

from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv=CountVectorizer(max_features=5000, stop_words='english')
```

```python
cv.fit_transform(df_movies_tags['tags']).toarray().shape  
```

```python
vectors=cv.fit_transform(df_movies_tags['tags']).toarray() #convert to vector
```

```python
vectors[0]
```

```python
len(cv.get_feature_names())
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return y
```

```python
df_movies_tags['tags']= df_movies_tags['tags'].apply(stem)
df_movies_tags
```

```python
#compares each and every movie to each and every other movie and computes the similarity between them
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape #4805 movies
```

```python
#storing the cosine similarity values in a variable
similarity=cosine_similarity(vectors)
similarity[1]
```

```python
#sorting the vector list in reverse because we want the top 5 ([1:6]) movies that are most similar to the specified movie (indicated by the index of 0)
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=df_movies_tags[df_movies_tags['title']==movie].index[0] #keeps track of the index of the movie that was entered by user
    #print(movie_index)
    distances=similarity[movie_index] #finding the cosine similarity values associated with the movie index
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6] #sort by most similar movies
    
    for i in movies_list:
        print(df_movies_tags.iloc[i[0]].title)
```

### Enter a movie below to get a recommendation

```python
recommend('The Dark Knight')
```
